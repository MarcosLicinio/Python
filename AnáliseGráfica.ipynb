```python
import pandas as pd

tabela = pd.read_csv("cancelamentos.csv")
tabela = tabela.drop("CustomerID", axis=1)
tabela = tabela.dropna()
display(tabela)
display(tabela.info())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
display(tabela.groupby("duracao_contrato").mean(numeric_only=True))
#taxa de cancelamentos na assinatura mensal é de 100%, porém a relação entre duração do contrato não demonstra tendências com outras variáveis.
tabela = tabela[tabela["duracao_contrato"]!="Monthly"]
#excluir os dados de assinatura mensal para analizar novamente, criar gráficos talvez seja melhor.
import plotly.express as px
for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou")
    grafico.show()
#provavelmente os dias de atraso, idade e ligações ao call center são o grande gargalo desta empresa.
```
