```python
import pandas as pd

tabela = pd.read_csv("cancelamentos.csv")
tabela = tabela.drop("CustomerID", axis=1)
tabela = tabela.dropna()
display(tabela)
display(tabela.info())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
display(tabela.groupby("duracao_contrato").mean(numeric_only=True))
#taxa de cancelamentos na assinatura mensal Ã© de 100%, porÃ©m a relaÃ§Ã£o entre duraÃ§Ã£o do contrato nÃ£o demonstra tendÃªncias com outras variÃ¡veis.
tabela = tabela[tabela["duracao_contrato"]!="Monthly"]
#excluir os dados de assinatura mensal para analizar novamente, criar grÃ¡ficos talvez seja melhor.
import plotly.express as px
for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou")
    grafico.show()
#provavelmente os dias de atraso, idade e ligaÃ§Ãµes ao call center sÃ£o o grande gargalo desta empresa.
```
